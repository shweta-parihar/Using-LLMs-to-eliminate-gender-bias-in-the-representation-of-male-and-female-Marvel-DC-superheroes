```python
import numpy as np

list_rows = df_male.values.tolist()
list_rows_2 = []
for i in list_rows:
  ds_list = [int(value) for value in i]
  list_rows_2.append(ds_list)
list_rows_2 = np.array(list_rows_2)
list_rows_2
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers




# df_ds = df_name[[col_name]]
# df_ds = df_ds.dropna()
# df_ds = df_ds[pd.to_numeric(df_ds[col_name], errors='coerce').notnull()]
# print(len(df_ds))
# ds_list = list(df_ds[col_name])
# ds_list = [int(value) for value in ds_list]
# ds_list = np.array(ds_list)


# Generate random data for training
input_size = 57
latent_dim = 50

# Generator model
def build_generator(latent_dim, output_size):
    model = tf.keras.Sequential()
    model.add(layers.Dense(256, input_dim=latent_dim, activation='relu'))
    model.add(layers.Dense(output_size, activation='relu'))
    return model

# Discriminator model
def build_discriminator(input_size):
    model = tf.keras.Sequential()
    model.add(layers.Dense(256, input_dim=input_size, activation='relu'))
    model.add(layers.Dense(1, activation='sigmoid'))
    return model

# GAN model
def build_gan(generator, discriminator):
    discriminator.trainable = False
    model = tf.keras.Sequential()
    model.add(generator)
    model.add(discriminator)
    return model

# Compile discriminator
discriminator = build_discriminator(input_size)
discriminator.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Compile GAN
generator = build_generator(latent_dim, input_size)
gan = build_gan(generator, discriminator)
gan.compile(loss='binary_crossentropy', optimizer='adam')

# Training GAN
epochs = 1000
batch_size = 64

for epoch in range(epochs):
    noise = np.random.normal(0, 1, size=(batch_size, latent_dim))
    generated_data = generator.predict(noise)

    # real_data = np.random.normal(0, 1, size=(batch_size, input_size))
    real_data = list_rows_2[np.random.randint(0, list_rows_2.shape[0], size=batch_size)]
    generated_data = np.array(generated_data)
    X = np.concatenate([real_data, generated_data])
    y_dis = np.zeros(2 * batch_size)
    y_dis[:batch_size] = 1

    # Train discriminator
    d_loss = discriminator.train_on_batch(X, y_dis)

    # Train generator
    noise = np.random.normal(0, 1, size=(batch_size, latent_dim))
    y_gen = np.ones(batch_size)
    g_loss = gan.train_on_batch(noise, y_gen)

    # Print progress
    if epoch % 100 == 0:
        print(f"Epoch {epoch}, Discriminator Loss: {d_loss[0]}, Generator Loss: {g_loss}")

```
